##Hiring Timeline:

How does the number of job postings vary by month? (Line chart of postings over time)

```python
# Assuming 'df' has a 'Posting Date' column with datetime values
df['Posting Date'] = pd.to_datetime(df['Posted Date'])  # Convert to datetime if not already

# Extract month and year from the posting date
df['Month'] = df['Posting Date'].dt.to_period('M')

# Group by month and count the number of job postings
monthly_postings = df.groupby('Month').size().reset_index(name='Job Postings')

# Convert 'Month' to string format for plotting
monthly_postings['Month'] = monthly_postings['Month'].astype(str)

# Set an attractive background style
sns.set_style("whitegrid")
plt.figure(figsize=(14, 8))  # Larger figure size for better visibility

# Create a line plot for job postings over time
ax = sns.lineplot(
    x='Month', 
    y='Job Postings', 
    data=monthly_postings, 
    marker='o', 
    color='b', 
    linewidth=2.5
)

# Add a bold and stylish title
plt.title("ðŸ“… Number of Job Postings by Month", fontsize=18, fontweight='bold', color='darkgreen')

# Label x and y axes with a clear font
plt.xlabel("Month", fontsize=14, fontweight='bold', color='darkred')
plt.ylabel("Number of Job Postings", fontsize=14, fontweight='bold', color='darkred')

# Rotate x-axis labels for readability
plt.xticks(rotation=45, ha='right', fontsize=12, color="darkblue")
ax.set_xticklabels(monthly_postings['Month'])

# Show gridlines to make the plot easier to read
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Show the plot
plt.show()
```

What is the average duration between job posting and application deadline? (Histogram)

```python

# Calculate the duration between job posting and application deadline
df['Application Deadline'] = pd.to_datetime(df['Application Deadline'])  # Convert to datetime if not already
df['Duration'] = (df['Application Deadline'] - df['Posting Date']).dt.days

# Set an attractive background style
sns.set_style("whitegrid")
plt.figure(figsize=(14, 8))  # Larger figure size for better visibility

# Create a histogram to show the distribution of durations
ax = sns.histplot(
    df['Duration'], 
    bins=30,  # Number of bins for the histogram
    kde=True,  # Add a kernel density estimate
    color='blue',  # Color of the bars
    edgecolor='black'  # Color of the edges
)

# Add a bold and stylish title
plt.title("â³ Distribution of Duration Between Job Posting and Application Deadline", fontsize=18, fontweight='bold', color='darkgreen')

# Label x and y axes with a clear font
plt.xlabel("Duration (Days)", fontsize=14, fontweight='bold', color='darkred')
plt.ylabel("Frequency", fontsize=14, fontweight='bold', color='darkred')

# Show gridlines to make the plot easier to read
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Show the plot
plt.show()

# Calculate and print the average duration
average_duration = df['Duration'].mean()
print(f"The average duration between job posting and application deadline is: {average_duration:.2f} days")
```
