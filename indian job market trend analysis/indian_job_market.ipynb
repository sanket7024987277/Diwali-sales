##Hiring Timeline:

What is the average duration between job posting and application deadline? (Histogram)

```python

# Calculate the duration between job posting and application deadline
df['Application Deadline'] = pd.to_datetime(df['Application Deadline'])  # Convert to datetime if not already
df['Duration'] = (df['Application Deadline'] - df['Posting Date']).dt.days

# Set an attractive background style
sns.set_style("whitegrid")
plt.figure(figsize=(14, 8))  

# Create a histogram to show the distribution of durations
ax = sns.histplot(
    df['Duration'], 
    bins=30,  
    kde=True,  
    color='blue', 
    edgecolor='black'  
)

# Add a bold and stylish title
plt.title("â³ Distribution of Duration Between Job Posting and Application Deadline", fontsize=18, fontweight='bold', color='darkgreen')

# Label x and y axes with a clear font
plt.xlabel("Duration (Days)", fontsize=14, fontweight='bold', color='darkred')
plt.ylabel("Frequency", fontsize=14, fontweight='bold', color='darkred')

# Show gridlines to make the plot easier to read
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Show the plot
plt.show()

# Calculate and print the average duration
average_duration = df['Duration'].mean()
print(f"The average duration between job posting and application deadline is: {average_duration:.2f} days")
```

CONCLUSION :-
Based on the above data analysis, we can conclude the following:

1. Job Market Trends: The most in-demand job roles are those with high frequency in job postings. Companies like Amazon, Infosys, and Deloitte are posting the most job openings, with a significant number of opportunities available in cities like Ahmadabad, Delhi, and Jaipur.

2. Salary and Experience Analysis: The distribution of salary ranges indicates that certain salary ranges have the highest number of job postings. The average duration between job posting and application deadline is approximately 18.56 days, suggesting a relatively short application window for job seekers.

3. Location Insights: Cities such as Ahmadabad, Delhi, and Jaipur have the highest number of job opportunities. Remote work opportunities are predominantly offered by companies like Amazon and Infosys, highlighting a trend towards flexible working arrangements.

4. Application Trends: Job portals like LinkedIn, Indeed, and Naukri.com are attracting the highest number of applicants. Job titles with the highest number of applications indicate high competition for these roles.

5. Company Analysis: The distribution of job postings by company size shows that large companies (500+ employees) are more active in hiring. Additionally, companies offering remote work opportunities are becoming increasingly popular among job seekers.

Overall, the analysis provides valuable insights into the current job market trends, salary expectations, and the importance of remote work opportunities. This information can help job seekers tailor their applications and focus on the most promising job roles and locations.
